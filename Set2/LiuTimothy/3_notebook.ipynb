```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import os
%matplotlib inline
```

## Load data

```python
train_file = os.getcwd() + '/data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]
```

# Lasso

```python
alphal = np.linspace(0.01, 3, 30)
wl = np.zeros((30, 9))    #weights for lasso regression

#loop running lasso classifier
for i in range(30):
    clfl = Lasso(alpha = alphal[i])
    clfl.fit(X_train, y_train)
    wl[i] = clfl.coef_  
    
```

```python
#plot lasso weights 
plt.figure('lasso')
colors = ['red', 'salmon', 'orange', 'gold', 'yellow', 'lawngreen', 'green', 'cyan', 'b']
for i in range(9):
    plt.scatter(alphal, wl[:, i], color = colors[i])
plt.legend(['w' + str(x + 1) for x in range(9)])
plt.grid(True)
plt.title('Lasso')
plt.xlabel("alpha")
plt.ylabel("Weight Value")
plt.savefig("lasso")
plt.show()
```

# Ridge

```python
alphar = np.linspace(1, 10000, 10000)
wr = np.zeros((10000, 9))  #weights for ridge regressions

#loop running ridge classifier
for i in range(10000):
    clfr = Ridge(alpha = alphar[i])
    clfr.fit(X_train, y_train)
    wr[i] = clfr.coef_

```

```python
#plot ridge weights 
colors = ['red', 'salmon', 'orange', 'gold', 'yellow', 'lawngreen', 'green', 'cyan', 'b']
plt.figure('ridge')
#print(wr)
for i in range(9):
    plt.scatter(alphar, wr[:, i], color = colors[i])
plt.legend(['w' + str(x + 1) for x in range(9)])
plt.grid(True)
plt.title('Ridge')
plt.xlabel("alpha")
plt.ylabel("Weight Value")
plt.xlim(-1000, 13000)
plt.ylim(-5, 5)
plt.savefig("ridge")
plt.show()
```
